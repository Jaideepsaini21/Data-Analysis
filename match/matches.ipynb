```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
match = pd.read_csv('matches.csv')
```

```python
match.head()
```

```python
plt.rcParams['figure.dpi']=100
plt.rcParams['figure.figsize']=10,5
```

```python
match['toss_winner'].value_counts()
```

```python
i1 = match['toss_winner'].value_counts().index
i2 = match['toss_winner'].value_counts().values
plt.bar(i1,i2)
plt.xticks(rotation=90)
plt.show()
```

```python
i1 = match['toss_winner'].value_counts().index
i2 = match['toss_winner'].value_counts().values
plt.bar(i1,i2)
plt.xticks(rotation=90)
plt.text('Sunrisers Hyderabad',40, 'hello')
plt.show()
```

```python
plt.bar(i1,i2)
plt.xticks(rotation=90)
plt.text(1,93, 'hello',c='red',rotation=45)
plt.show()
```

```python
plt.bar(i1, i2)
plt.xticks(rotation=90)
for i in range(len(i1)):
    plt.text(i,i2[i]+5, i2[i], ha='right', va='center')
```
